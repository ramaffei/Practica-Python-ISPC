Ejercicio 2.1.

Escribir un programa que le pregunte al usuario una cantidad de pesos, una tasa de interés y un número de años y muestre como resultado el monto final a obtener. La fórmula a utilizar es:

Cn = C * (1 + x/100) ^ n

Donde C es el capital inicial, x es la tasa de interés y n es el número de años a calcular.


```python
print("Bienvenido")
pesos = float(input("Ingresa la cantidad de pesos que deseas depositar: "))
interes = float(input("Ingresa el interes anual: "))
anos = int(input("Ingresa cuantos años dejaras la inversion: "))

Cn = pesos * (1 + interes/100) ** anos

print("Recibiras $"+str(Cn)+" al cabo de "+str(anos)+" años con un interes del "+str(interes)+"%")
```

Ejercicio 2.2.

Escribir un programa que convierta un valor dado en grados Fahrenheit a grados Celsius. Recordar que la fórmula para la conversión es:

F = 9/5 * C + 32.

```python
F = input("Ingresa la cantidad de grados en Fahrenheit")
C = (float(F)-32) * 5/9

print(F,"°F son:",str(C),"°C")
```

Ejercicio 2.6.

Los alumnos de un curso se han dividido en dos grupos A y B de acuerdo al sexo y el nombre. El grupo A esta formado por las mujeres con un nombre anterior a la M y los hombres con un nombre posterior a la N y el grupo B por el resto. Escribir un programa que pregunte al usuario su nombre y sexo, y muestre por pantalla el grupo que le corresponde.

```python
name = input("¿Cómo te llamas? ")
gender = input("¿Cuál es tu sexo (M o F)? ")
if gender == "F":
    if name.lower() < "m":
        group = "A"
    else:
        group = "B"
else:
    if name.lower() > "n":
        group = "A"
    else:
        group = "B"
print("Tu grupo es " + group)
```

Escribir un programa que imprima todos los números pares entre dos números que se le pidan al usuario.

Ejercicio 2.8.

Escribir un programa que pida al usuario un número entero positivo y muestre por pantalla la cuenta desde atrás de ese número hasta cero separados por comas.

```python
num = int(input("Ingrese un numero para comenzar la cuenta regresiva"))
while num > 0:
    print(num,end=",")
    num-=1
print(num)
```

Ejercicio 2.9.

Escribir un programa que pida al usuario un número entero positivo y muestre por pantalla todos los números impares desde 1 hasta ese número separados por comas. Ayuda: usar el módulo de 2 (%)

```python
num = int(input("Ingrese un numero para comenzar"))
for i in range(1,num+1):
    if i%2!=0:
        print(i, end=",")
```
