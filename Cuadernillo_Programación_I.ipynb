1. Introducir los lados de un triángulo y visualizar por pantalla si dicho triángulo es equilátero, isósceles o escaleno.

```python
lado1 = float(input("Ingresa la longitud del primer lado"))
lado2 = float(input("Ingresa la longitud del segundo lado"))
lado3 = float(input("Ingresa la longitud del tercer lado"))

if lado1 == lado2:
    if lado1 == lado3:
        print("Es un triangulo equilatero")
    else:
        print("Es un Triángulo isósceles")
elif lado1 == lado3 or lado2 == lado3:
    print("Es un Triángulo isósceles")
else:
    print("Es un Triángulo escaleno")
```

Ejercicios estructuras repetitivas y estructuras condicional.

1. Realice un programa que lea 4 números y diga cuantos son pares y cuantos impares y devuelva la
sumatoria de los pares.

```python
num1 = int(input("Ingresa un numero"))
num2 = int(input("Ingresa un numero"))
num3 = int(input("Ingresa un numero"))
num4 = int(input("Ingresa un numero"))

pares = 0
impares = 0
suma = 0

for i in [num1, num2, num3, num4]:

    if i%2 == 0:
        pares += 1
        suma += i
    else:
        impares += 1

print(pares, "numeros pares")
print(impares, "numeros impares")
print(suma, "suma de numeros pares")





    
```

2. Leer 10 números y obtener la cantidad de mayores y la cantidad de menores a 100, cuál es el número
máximo y cuál es el número mínimo.

```python
mayora100, menora100, maximo, minimo, i = 0,0,0,0,0

while i<10:
    numero = int(input("Ingresa un numero:"))

    if numero>100:
        mayora100 += 1
    else:
        menora100 += 1
    
    if numero>maximo:
        maximo = numero
    
    if numero<minimo or minimo == 0:
        minimo = numero

    i += 1

    print(numero,end=",")

print("\nNumeros mayores a 100:", mayora100)
print("Numeros menores a 100:", menora100)
print("Numero maximo:", maximo)
print("Numero minimo:", minimo)
```

3. Ingresar las edades y el sexo de 15 personas y determinar cuántas son mujeres, cuántos varones,
cuántos mayores de edad y cuántos menores de edad.

```python
mujeres, varones, mayores, menores = 0, 0, 0, 0

for i in range(0, 15):
    edad = int(input("Ingresa tu edad:"))
    sexo = input("Ingresa M para masculino, F para femenino")

    print(i+1,": edad:",edad,"- sexo:", sexo.upper())

    if sexo.upper() == 'F':
        mujeres += 1
    elif sexo.upper() == 'M':
        varones += 1
    
    if edad >= 18:
        mayores += 1
    else:
        menores += 1

print("Mujeres:", mujeres)
print("Varones:", varones)
print("Mayores:", mayores)
print("Menores:", menores)
```
